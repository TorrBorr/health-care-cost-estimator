Need full state name to create County_state variable. This cool function will convert abbreviations to full names: 

```python
import us

def abb2state(name, convert=False):
    state_list = {'from_field': 'name', 'to_field': 'abbr'} if convert else {'from_field': 'abbr', 'to_field': 'name'}
    states = pd.DataFrame(us.states.mapping(**state_list), index=['Name']).transpose()
    name = pd.Series(name)

    def single_a2s(s):
        return states.loc[s]['Name'].encode('utf8') if s in states.index else None

    return name.apply(single_a2s).tolist()
```
